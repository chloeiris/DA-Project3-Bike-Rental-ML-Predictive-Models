```python
import pandas as pd
import numpy as np
import os
import datetime as dt
```

#### En este jupyter vamos a modificar la fecha para que siga un formato igual con nuestros datos usados para los modelos,y así poder unir datos también

```python
# Cargamos los tres dataframes para modificar y unificarlos
trips = pd.read_csv('../Datos/4. nuevo_data_trips/trips_bikes2019.csv', index_col=0)
```

```python
trips.head(2)
```

```python
bikes = pd.read_pickle('../Datos/2. data_limpia_para_modelar/bikes_eda_limpio.pkl')
bikes.head(2)
```

```python
stations = pd.read_csv('../Datos/4. nuevo_data_trips/stations_info.csv', index_col=0)
stations.head(2)
```

```python
# Hacemos una copia del que vamos a modificar
tripscopy = trips.copy()
```

```python
# Separamos fecha de hora creando una nueva columna con la hora
tripscopy['Start time'] = tripscopy['Start date'].apply(lambda x: x.split(' ')[1])
```

```python
tripscopy['End time'] = tripscopy['End date'].apply(lambda x: x.split(' ')[1])
```

```python
tripscopy.head()
```

```python
# Cambiamos el dtype a datetime para poder modificar el formato
tripscopy['Start date'] = pd.to_datetime(tripscopy['Start date'])
```

```python
tripscopy.head(2)
```

```python
# Definimos una funcion para darle nuevo formato que se ajuste con nuestro dataset bikes
def date_format(x):

    return dt.datetime.strftime(x, '%d-%m-%Y')
```

```python
tripscopy['Start date'] = tripscopy['Start date'].apply(date_format)
```

```python
tripscopy.head(2)
```

```python
tripscopy['End date'] = pd.to_datetime(tripscopy['End date'])
```

```python
tripscopy['End date'] = tripscopy['End date'].apply(date_format)
```

```python
tripscopy.head(2)
```

```python
# Comprobamos que las fechas están bien
tripscopy['Start date'].unique()
```

```python
# Para unir con el archivo bikes, tenemos que quedarnos con una fecha, pero para ello tendremos que decidir si elegimos la fecha de comienzo o de fin. Hay 30.743 regitros que empiezan en un día y acaban en el siguiente
tripscopy[tripscopy['Start date'] != tripscopy['End date']]
```

```python
long = tripscopy[tripscopy['Start date'] != tripscopy['End date']]
```

```python
# De estos viajes, cual es su duracion?
long['Duration'].max()
```
