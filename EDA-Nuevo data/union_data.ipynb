```python
import os
import pandas as pd
import numpy as np
```

#### En este jupyter vamos a unificar los datos de alquiler de bicicletas obtenidos de Capital Share Bikes. Cada csv pertenece a un mes distinto, y tenemos los datos para los años 2018 y 2019.

```python
os.getcwd()
```

```python
os.chdir('../Datos/4. nuevo_data_trips/2019')
```

```python
# Los arichivos a unificar
archivos = os.listdir()
archivos
```

```python
# Diccionario vacío para ir apendeando los csv en dataframes
fechas_dict = {}
```

```python
# Prueba para insertar dataframes
fechas_dict[archivos[0]] = pd.read_csv(archivos[0], index_col=0)
fechas_dict[archivos[0]]
```

```python
# Inserción de dataframes
for i in archivos:
    fechas_dict[i] = pd.read_csv(i, index_col=0)
```

```python
# Deberiamos tener tantos items en el diccionario como csvs a unir
len(archivos) == len(fechas_dict)
```

```python
# Extraemos los dataframes a una lista
df_list = [ v for k,v in fechas_dict.items()] 
df_list
```

```python
# Concatenamos los df para unificar la información
trips_info = pd.concat(df_list, axis=0)
```

```python
trips_info.head(5)
```

```python
# Vemos si tiene sentido el tamaño del nuevo dataframe
trips_info.shape
```

```python
# Reseteamos el índice
trips_info.reset_index(inplace=True)
```

```python
trips_info.head()
```

```python
# Guardamos a csv
trips_info.to_csv('../trips_bikes2019.csv')
```

#### Extracción del json stations_information a csv

```python
station_info = pd.read_json('../station_information.json')
station_info
```

```python
station_info2 = station_info['data'].apply(pd.Series)
station_info2.reset_index(inplace=True, drop=True)
station_info2.head()
```

```python
stations = station_info2.T
stations
```

```python
stations = stations[0].apply(pd.Series)
stations.head()
```

```python
# Nos quedamos únicamente con las columnas que nos interesan
stations= stations[['short_name', 'capacity', 'name', 'region_id', 'lat', 'lon']]
stations
```

```python
# Guardamos como csv
stations.to_csv('../stations_info.csv')
```
